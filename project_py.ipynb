```python
import pandas as pd
import numpy as np

df=pd.read_csv("uncleaned_dataset_100_rows.csv")

df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
df.loc[df["Age"] < 0, "Age"] = np.nan
df.loc[df["Age"] > 100, "Age"] = np.nan
df["Age"] = df["Age"].fillna(25)


df["Name"] = df["Name"].fillna("Unknown")
df.loc[df["Name"] == "25", "Name"] = "Unknown"


df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce")
df.loc[df["Marks"] < 0, "Marks"] = np.nan
df.loc[df["Marks"] > 100, "Marks"] = np.nan
df["Marks"] = df["Marks"].fillna(df["Marks"])


df=df.to_string()
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

Age_gap = df["Age"].value_counts()


print("Age gap:")
print(Age_gap)


plt.figure(figsize=(10, 6))
bars = plt.bar(Age_gap.index, Age_gap.values, color='skyblue')


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height,
             f'{height}', ha='center', va='bottom')


plt.xlabel('male')
plt.ylabel('female')
plt.title('Count of Each Gender')
plt.tight_layout()


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("uncleaned_dataset_100_rows.csv")


df["Marks"] = pd.to_numeric(df["Marks"], errors="coerce").fillna(0)


def category(mark):
    if mark >= 75:
        return "High"
    elif mark >= 50:
        return "Medium"
    else:
        return "Low"

df["Category"] = df["Marks"].apply(category)


counts = df["Category"].value_counts()


plt.figure(figsize=(6,6))
plt.pie(
    counts,
    labels=counts.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Student Marks Category")
plt.axis("equal")
plt.show()
```
